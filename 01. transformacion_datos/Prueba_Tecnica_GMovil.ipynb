```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dset = pd.read_csv("C:/Users/luis-/Documents/data/data_org.csv", sep=';', encoding = "ISO-8859-1")
dset.head()
```

```python
dset.columns
```

```python
fecha = dset["Fecha Clearing"].str.split(expand=True)
fecha.columns = ['FechaContable', 'HoraContable']
```

```python
parada = dset["Parada"].str.split('[)] ',expand=True)
parada.columns = ['Paradero', 'StrParadero']
linea = dset["Linea"].str.split(expand=True)
linea.columns = ['Linea_aux', 'StrLinea']
dset = pd.concat([dset, fecha, parada, linea], axis=1)
dset = dset.drop(['Fecha Clearing', 'HoraContable', "Parada","Linea_aux","Linea","Tipo Vehiculo"], axis = 1)
dset.rename(columns={'Día Trx':'FechaTransaccion','Hora Trx':'HoraTransaccion','Operador':'Empresa','Ruta Modificada':'RutaSae','Linea SAE':'Linea','Vehiculo':'Bus','Tipo de Viaje':'TipoValidacion'},inplace=True)
dset["Paradero"] = dset["Paradero"].str.replace(r"\W","")
parada2 = dset["StrParadero"].str.split('_', 1, expand=True)
parada2.columns = ['Cenefa', '0']
dset = pd.concat([dset, parada2], axis=1)
dset = dset.drop(['0'], axis = 1)
dset["StrRutaSae"]= ""
dset = dset[['FechaContable', 'FechaTransaccion', 'HoraTransaccion', 'Empresa', 'Paradero', 'StrParadero', 'Linea', 'StrLinea', 'Bus', 'TipoValidacion', 'RutaSae', 'StrRutaSae', 'Cenefa']]
dset
```

```python
dset.to_csv("C:/Users/luis-/Documents/data/data_fnl.csv", index=True)
```
